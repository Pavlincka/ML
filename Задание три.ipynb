# Поиск выбросов и аномалий

**Те, кто выполняет задания со * делают поиск выбросов с помощью написанного ранее класса**

Найти выбросы по числовой переменной (использовать метод основанный на среднем и на межквартильном размахе). Объясните результат

Вывести статистику с выбросами и без. Сравните и объясните результат


```python
#Методом среднего

# Функция для определения выбросов методом среднего
def detect_outliers_mean(dataframe, column_name, threshold=3):
    """
    Функция для определения выбросов методом среднего
    Args:
        dataframe: DataFrame, исходный DataFrame
        column_name: str, имя столбца с числовой переменной
        threshold: float, пороговое значение для определения выбросов (по умолчанию: 3)
    Returns:
        DataFrame без выбросов
    """
    mean = np.mean(dataframe[column_name])
    std = np.std(dataframe[column_name])
    lower_bound = mean - threshold * std
    upper_bound = mean + threshold * std
    df_no_outliers = dataframe[(dataframe[column_name] >= lower_bound) & (dataframe[column_name] <= upper_bound)]
    return df_no_outliers

# Вызов функции для поиска и удаления выбросов
df_no_outliers = detect_outliers_mean(df, 'RestingBP')

# Вывод DataFrame без выбросов
df_no_outliers
```

```python
#При помощи межквартильного размаха

def remove_outliers(df, column):
    # Рассчитываем первый и третий квартили
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    
    # Рассчитываем межквартильный размах
    iqr = q3 - q1
    
    # Определяем границы для определения выбросов
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Фильтруем датафрейм, оставляя только значения внутри границ
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df_filtered

# Удаляем выбросы по переменной
df_filtered = remove_outliers(df, 'RestingBP')

df_filtered
```

```python
#Статистика и объяснение

print(f'1. Оригинальный\n{df.describe()}\n\n2. Метод среднего\n{df_no_outliers.describe()}\n\n3. Межквартильный размах\n{df_filtered.describe()}')
```

**Объяснения**

Итак, оригинальный датафрейм содержит некоторые выбросы, но для анализа сводной статистики, были использованы два метода на основе критерия среднего и межквартильного размаха для получения более точных и надежных результатов. В исходном наборе данных у нас есть 918 наблюдений, тогда как в методе среднего и межквартильном размахе мы исключили некоторые выбросы, получив результирующие наборы данных с 910 и 890 наблюдениями соответственно.
