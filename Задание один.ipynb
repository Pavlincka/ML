```python
intArr =[]
objArr = []
for i in df.columns:
    if df[i].dtype == 'int64':
        intArr.append(i)
    else:
        objArr.append(i)

df[intArr].min()
```

```python
df[intArr].max()
```

Типизация
Присвойте соответствующим колонкам один из типов (category, float, int)

```python
for i in intArr:
    df[i] = df[i].astype('Int64')

for obj in objArr:
    df[obj] = df[obj].astype('category')

df.dtypes
```

Ответ: числовые количественные (Population, Violent crime), категориальные (City)
