```python
intArr =[]
objArr = []
for i in df.columns:
    if df[i].dtype == 'int64':
        intArr.append(i)
    elif df[i].dtype == 'float64':
        intArr.append(i)
    else:
        objArr.append(i)

df[intArr].min()
```

```python
df[intArr].max()
```

Типизация
Присвойте соответствующим колонкам один из типов (category, float, int)

```python
for obj in objArr:
    df[obj] = df[obj].astype('category')

df.dtypes
```
